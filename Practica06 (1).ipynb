## 1. Contar del 1 al 10
Escribe un programa que use un ciclo `while` para imprimir los números del 1 al 10.

```python
# Escribe tu solución aquí
val1=1
while val1<11:
  print(val1)
  val1=val1+1
```

## 2. Pedir una contraseña hasta que sea correcta
Pide al usuario que escriba una contraseña. Mientras no escriba `python123`, el programa debe seguir preguntando.

```python
# Escribe tu solución aquí
v1=input("Dame una contraseña")
while v1 != "python123":
  print("Contraseña incorrecta.Intenta de nuevo.")
  v1 = input("Dame una contraseña")
  print("¡Contraseña correcta;!")
```

## 3. Sumar números mientras el usuario no escriba 0
Pide números al usuario y súmalos. El ciclo termina cuando el usuario introduce 0. Al final muestra la suma total.

```python
# Escribe tu solución aquí
var1 = int(input("Dame un numero:"))
var2 = int(input("Dame otro numero:"))
while(var1+var2 != 0):
    var2=int(input("Dame un numero:"))
    var1=int(input("Dame otro numero:"))
```

## 4. Contador regresivo
Pide al usuario un número entero positivo y usa un ciclo `while` para imprimir una cuenta regresiva hasta llegar a 0.

```python
# Escribe tu solución aquí

var1 = int(input("Dame un numero positivo"))
while(var1>0):
    print(var1)
    var1=var1-1
```
